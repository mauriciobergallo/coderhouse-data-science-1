## 2. y 3. Crear grÃ¡ficos que usen al menos tres variables y hacer un diagnÃ³stico de lo que los mismos nos estÃ¡n diciendo

```python
# Obtenemos un subset de paises con mas y menos libertades econÃ³micas
# Creamos y comparÃ¡mos un grÃ¡fico de precios de comida
df_filtered_by_country = df_cost_of_living.query("country == 'Argentina' | country == 'Venezuela' | country == 'Bolivia' | country == 'Uruguay' | country == 'Chile' | country == 'Peru'")
df_grouped_by_country =  df_filtered_by_country[['country', 'meal_inexpensive_restaurant', 'mc_meal']].groupby('country').mean()

countries = ['Argentina', 'Bolivia', 'Chile', 'Peru', 'Uruguay', 'Venezuela']
products_means = {
    'McMeal': df_grouped_by_country['mc_meal'].to_numpy(),
    'Comida en Restaurante': df_grouped_by_country['meal_inexpensive_restaurant'].to_numpy(),
}

x = np.arange(len(countries))  # the label locations
width = 0.35  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in products_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('USD')
ax.set_title('Valor promedio por paÃ­ses')
ax.set_xticks(x + width, countries)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 15)

plt.show()
```

- Hipotesis #1: podemos ver en los grÃ¡ficos que los costos de restaurante son menores a los de un McMeal debido a que mantienen un precio internacional.

```python
# Creamos un grÃ¡fico con precios de comida generalmente conocidos como canasta bÃ¡sica
df_filtered_by_country = df_cost_of_living.query("country == 'Argentina' | country == 'Venezuela' | country == 'Bolivia' | country == 'Uruguay' | country == 'Chile' | country == 'Peru'")
df_grouped_by_country =  df_filtered_by_country[['country', 'rice_white_1000', 'milk_regular_1']].groupby('country').mean()

countries = ['Argentina', 'Bolivia', 'Chile', 'Peru', 'Uruguay', 'Venezuela']
products_means = {
    'Arroz 1 Kg': df_grouped_by_country['rice_white_1000'].to_numpy(),
    'Leche 1 L': df_grouped_by_country['milk_regular_1'].to_numpy(),
}

x = np.arange(len(countries))  # the label locations
width = 0.35  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in products_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('USD')
ax.set_title('Valor promedio por paÃ­ses')
ax.set_xticks(x + width, countries)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 2)

plt.show()
```

- Hipotesis 2: Como se ve en los grÃ¡ficos, los paises con menores libertades econÃ³micas suelen tener controles de precios sobre los productos conocidos como "Canasta BÃ¡sica" en este caso, Leche y Arroz. Por ende estos precios se ven levementa mas bajos. Hay una mensiÃ³n especial con Venezuela, que los costos de estos productos son mas altos que en el resto. TendrÃ­amos que buscar mas datos para entender esta situaciÃ³n.

```python
# Creamos un grÃ¡fico con precios de ropa, para poder comparar

# df_filtered_by_country = df_cost_of_living.query("country == 'Argentina' | country == 'Venezuela' | country == 'Bolivia' | country == 'Uruguay' | country == 'Chile' | country == 'Peru'")
df_grouped_by_country =  df_filtered_by_country[['country', '1_pair_jeans', '1_pair_nike_running_shoes']].groupby('country').mean()

countries = ['Argentina', 'Bolivia', 'Chile', 'Peru', 'Uruguay', 'Venezuela']
products_means = {
    'Jeans': df_grouped_by_country['1_pair_jeans'].to_numpy(),
    'Zapatillas Nike Running': df_grouped_by_country['1_pair_nike_running_shoes'].to_numpy(),
}
width = 0.8  # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
bottom = np.zeros(6)

for sex, product_name in products_means.items():
    p = ax.bar(countries, product_name, width, label=sex, bottom=bottom)
    bottom += product_name

    ax.bar_label(p, label_type='center')

ax.set_title('ComparaciÃ³n de Precios de productos "Importados"')
ax.legend()

plt.show()
```
