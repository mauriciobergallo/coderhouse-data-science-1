```python
import pandas as pd
import matplotlib as mtp
import matplotlib.pyplot as plt
```

```python
# Describe the DataFrame
df_titanic = pd.read_csv('./datasets/titanic.csv')
df_titanic.describe()
```

```python
# Replace the column names, and the values for Passenger Class
df_titanic.rename(columns={'Pclass': 'PassengerClass', 'SibSp': 'SiblingsNumber', 'Parch': 'ParentChildrenNumber'}, inplace=True)
df_titanic['PassengerClass'].replace(1, 'First Class', inplace=True)
df_titanic['PassengerClass'].replace(2, 'Second Class', inplace=True)
df_titanic['PassengerClass'].replace(3, 'Third Class', inplace=True)

df_titanic.describe()
```

```python
# Show a Pie Chart showing the Survivors by Class
df_survivors = df_titanic.loc[df_titanic['Survived'] == 1]

passenger_class_survivors_list = df_survivors['PassengerClass'].unique()
first_class_survivors_count = df_survivors['PassengerClass'].value_counts()['First Class']
second_class_survivors_count = df_survivors['PassengerClass'].value_counts()['Second Class']
third_class_survivors_count = df_survivors['PassengerClass'].value_counts()['Third Class']

pie_survivors_values = [first_class_survivors_count, second_class_survivors_count, third_class_survivors_count]

pie_survivors, ax_pie_survivors = plt.subplots()
ax_pie_survivors.pie(pie_survivors_values, labels=passenger_class_survivors_list, autopct='%1.1f%%')
```

```python
# Deceased by Class
df_deceased = df_titanic.loc[df_titanic['Survived'] == 0]
passenger_class_deceased_list = df_deceased['PassengerClass'].unique()
first_class_deceased_count = df_deceased['PassengerClass'].value_counts()['First Class']
second_class_deceased_count = df_deceased['PassengerClass'].value_counts()['Second Class']
third_class_deceased_count = df_deceased['PassengerClass'].value_counts()['Third Class']

pie_deceased_values = [first_class_deceased_count, second_class_deceased_count, third_class_deceased_count]

pie_survivors, ax_pie_survivors = plt.subplots()
ax_pie_survivors.pie(pie_deceased_values, labels=passenger_class_deceased_list, autopct='%1.1f%%')
```
