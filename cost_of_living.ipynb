## 2. y 3. Crear grÃ¡ficos que usen al menos tres variables y hacer un diagnÃ³stico de lo que los mismos nos estÃ¡n diciendo

```python
# Obtenemos un subset de paises con mas y menos libertades econÃ³micas
df_countries_more_liberals = df_cost_of_living.query("country == 'Uruguay' | country == 'Chile' | country == 'Paraguay'")
df_countries_less_liberals = df_cost_of_living.query("country == 'Argentina' | country == 'Venezuela' | country == 'Bolivia'")

# Creamos y comparÃ¡mos un grÃ¡fico de precios de comida
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

sns.lineplot([df_countries_more_liberals['meal_inexpensive_restaurant'],
              df_countries_more_liberals['mc_meal']], ax=ax[0], markers=True)

sns.lineplot([df_countries_less_liberals['meal_inexpensive_restaurant'],
              df_countries_less_liberals['mc_meal']], ax=ax[1], markers=True)
```

- Hipotesis #1: podemos ver en los grÃ¡ficos que los costos de restaurante son menores a los de un McMeal debido a que mantienen un precio internacional.

```python
# Creamos un grÃ¡fico con precios de comida generalmente conocidos como canasta bÃ¡sica

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

sns.lineplot([df_countries_more_liberals['milk_regular_1'],
              df_countries_more_liberals['rice_white_1000']], ax=ax[0], markers=True)

sns.lineplot([df_countries_less_liberals['milk_regular_1'],
              df_countries_less_liberals['rice_white_1000']], ax=ax[1], markers=True)
```

- Hipotesis 2: Como se ve en los grÃ¡ficos, los paises con menores libertades econÃ³micas suelen tener controles de precios sobre los productos conocidos como "Canasta BÃ¡sica" en este caso, Leche y Arroz.

```python
# Creamos un grÃ¡fico con precios de ropa, para poder comparar

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

sns.lineplot([df_countries_more_liberals['1_pair_jeans'],
              df_countries_more_liberals['1_pair_nike_running_shoes']], ax=ax[0], markers=True)

sns.lineplot([df_countries_less_liberals['1_pair_jeans'],
              df_countries_less_liberals['1_pair_nike_running_shoes']], ax=ax[1], markers=True)
```

- Hipotesis #3: En contraposiciÃ³n los paÃ­ses con mayores libertades econÃ³micas suelen tener valores "mas bajos" con respecto al otro tipo de paÃ­ses ya que tienen menor cantidad de impuestos sobre este tipo de productos "importados".
