## 2. y 3. Crear gráficos que usen al menos tres variables y hacer un diagnóstico de lo que los mismos nos están diciendo

```python
# Obtenemos un subset de paises con mas y menos libertades económicas
df_countries_more_liberals = df_cost_of_living.query("country == 'Uruguay' | country == 'Chile' | country == 'Paraguay'")
df_countries_less_liberals = df_cost_of_living.query("country == 'Argentina' | country == 'Venezuela' | country == 'Bolivia'")

# Creamos y comparámos un gráfico de precios de comida
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

sns.lineplot([df_countries_more_liberals['meal_inexpensive_restaurant'],
              df_countries_more_liberals['mc_meal']], ax=ax[0], markers=True)

sns.lineplot([df_countries_less_liberals['meal_inexpensive_restaurant'],
              df_countries_less_liberals['mc_meal']], ax=ax[1], markers=True)
```

- Hipotesis #1: podemos ver en los gráficos que los costos de restaurante son menores a los de un McMeal debido a que mantienen un precio internacional.

```python
# Creamos un gráfico con precios de comida generalmente conocidos como canasta básica

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

sns.lineplot([df_countries_more_liberals['milk_regular_1'],
              df_countries_more_liberals['rice_white_1000']], ax=ax[0], markers=True)

sns.lineplot([df_countries_less_liberals['milk_regular_1'],
              df_countries_less_liberals['rice_white_1000']], ax=ax[1], markers=True)
```

- Hipotesis 2: Como se ve en los gráficos, los paises con menores libertades económicas suelen tener controles de precios sobre los productos conocidos como "Canasta Básica" en este caso, Leche y Arroz.

```python
# Creamos un gráfico con precios de ropa, para poder comparar

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

sns.lineplot([df_countries_more_liberals['1_pair_jeans'],
              df_countries_more_liberals['1_pair_nike_running_shoes']], ax=ax[0], markers=True)

sns.lineplot([df_countries_less_liberals['1_pair_jeans'],
              df_countries_less_liberals['1_pair_nike_running_shoes']], ax=ax[1], markers=True)
```

- Hipotesis #3: En contraposición los países con mayores libertades económicas suelen tener valores "mas bajos" con respecto al otro tipo de países ya que tienen menor cantidad de impuestos sobre este tipo de productos "importados".
