## 2. Crear gráficos que usen al menos tres variables y hacer un diagnóstico de lo que los mismos nos están diciendo

```python
# Creamos un gráfico de sueldos
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(20, 6)

salaries_in_southamerica_countries_rich = df_cost_of_living.query("country == 'Uruguay' | country == 'Chile' | country == 'Paraguay'")
sns.lineplot([salaries_in_southamerica_countries_rich['meal_2_ppl_mid_range_restaurant_three_courses'],
              salaries_in_southamerica_countries_rich['meal_inexpensive_restaurant'],
              salaries_in_southamerica_countries_rich['mc_meal']], ax=ax[0])

# valores para paises con "poca libertad económica"
salaries_in_southamerica_countries = df_cost_of_living.query("country == 'Argentina' | country == 'Venezuela' | country == 'Bolivia'")
sns.lineplot(data=[salaries_in_southamerica_countries['meal_2_ppl_mid_range_restaurant_three_courses'],
              salaries_in_southamerica_countries['meal_inexpensive_restaurant'],
              salaries_in_southamerica_countries['mc_meal']], ax=ax[1])

# plt.show()
# fig.show()
```
